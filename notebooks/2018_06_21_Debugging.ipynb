```python
from src.parameters import ANIMALS, SAMPLING_FREQUENCY

epoch_key = ('JZ1', 1, 2)
```

```python
from loren_frank_data_processing import get_interpolated_position_dataframe
from src.analysis import detect_epoch_ripples

position_info = get_interpolated_position_dataframe(epoch_key, ANIMALS)
ripple_times = detect_epoch_ripples(
    epoch_key, ANIMALS, SAMPLING_FREQUENCY, position_info)
```

```python
from loren_frank_data_processing import make_tetrode_dataframe

tetrode_info = make_tetrode_dataframe(ANIMALS).xs(
    epoch_key, drop_level=False)
```

```python
brain_areas = ['mec']
```

```python
is_brain_areas = tetrode_info.area.isin(brain_areas)
brain_areas_tetrodes = tetrode_info[is_brain_areas]
```

```python
from loren_frank_data_processing import get_multiunit_indicator_dataframe

mark_names = get_multiunit_indicator_dataframe(
    brain_areas_tetrodes.nchans.argmax(), ANIMALS).columns.tolist()
```

```python
marks = [(get_multiunit_indicator_dataframe(tetrode_key, ANIMALS)
          .loc[:, mark_names])
         for tetrode_key in brain_areas_tetrodes.index]
```

```python
position_info['lagged_linear_distance'] = (
    position_info.linear_distance.shift(1))
KEEP_COLUMNS = ['linear_distance', 'lagged_linear_distance', 'task',
                'is_correct', 'turn', 'speed', 'head_direction']
```

```python
from src.analysis import get_ripple_indicator

ripple_indicator = get_ripple_indicator(epoch_key, ANIMALS, ripple_times)
train_position_info = position_info[KEEP_COLUMNS].loc[
    ~ripple_indicator & position_info.is_correct].dropna()
```

```python

training_marks = np.stack([
    tetrode_marks.loc[train_position_info.index, mark_names]
    for tetrode_marks in marks], axis=0)
```

```python
from replay_classification import ClusterlessDecoder

n_position_bins=61
place_std_deviation=None
mark_std_deviation=20
confidence_threshold=0.8

decoder = ClusterlessDecoder(
    position=train_position_info.linear_distance.values,
    lagged_position=train_position_info.lagged_linear_distance.values,
    trajectory_direction=train_position_info.task.values,
    spike_marks=training_marks,
    n_position_bins=n_position_bins,
    place_std_deviation=place_std_deviation,
    mark_std_deviation=mark_std_deviation,
    replay_speedup_factor=16,
    confidence_threshold=confidence_threshold,
).fit()
```

```python
from src.analysis import _get_ripple_marks

test_marks = _get_ripple_marks(marks, ripple_times, SAMPLING_FREQUENCY)
results = [decoder.predict(ripple_marks, time.total_seconds())
           for ripple_marks, time in test_marks]
```

```python
from src.analysis import summarize_replay_results

(replay_info, decision_state_probability,
            posterior_density) = summarize_replay_results(
        results, ripple_times, position_info, epoch_key)
```

```python
from ipywidgets import interact

def _plot(ripple_number):
    g = posterior_density.isel(concat_dim=ripple_number).dropna('time').plot(
        x='time', y='position', col='state', col_wrap=2, robust=True, size=3)
    for ax in g.axes.flat:
        ax.axhline(replay_info.iloc[ripple_number].linear_distance,
                   color='white', linestyle='--', linewidth=3)
        
interact(_plot, ripple_number=(replay_info.index.min(), replay_info.index.max()))
```

```python
from scipy.io import loadmat
from loren_frank_data_processing.core import get_data_filename




animals = ANIMALS.copy()
multiunit_key = tetrode_info.index[0]


def get_multiunit_dataframe2(tetrode_key, animals):
    '''Retrieve the multiunits for each tetrode given a tetrode key
    This function retrieves data according to Demetris's format.

    Parameters
    ----------
    tetrode_key : tuple
        Unique key identifying the tetrode. Elements are
        (animal_short_name, day, epoch, tetrode_number).
    animals : dict of named-tuples
        Dictionary containing information about the directory for each
        animal. The key is the animal_short_name.

    Returns
    -------
    multiunit_dataframe : pandas dataframe
        The dataframe index is the time at which the multiunit occurred
        (in seconds). THe other values are values that can be used as the
        multiunits.

    '''
    animal, day, epoch, tetrode_number = tetrode_key
    try:
        multiunit_file = loadmat(
            get_data_filename(animals[animal], day, 'marks'))
        multiunit_data = multiunit_file['marks'][0, -1][0, epoch - 1][
                    0, tetrode_number - 1]

        time = pd.TimedeltaIndex(multiunit_data['times'][0, 0].squeeze(),
                                 unit='s', name='time')
        multiunit = multiunit_data['marks'][0, 0].astype(np.float)
        column_names = ['channel_{number + 1}_max'.format(number=number)
                        for number in np.arange(multiunit.shape[1])]
        return pd.DataFrame(multiunit, columns=column_names, index=time)
    except (FileNotFoundError, TypeError):
        print(get_data_filename(animals[animal], day, 'marks'))
```
