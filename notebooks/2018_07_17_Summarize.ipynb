```python
from loren_frank_data_processing import get_LFPs
from spectral_connectivity import Multitaper, Connectivity
from src.analysis import _center_time

def get_power(tetrode_info, animals, multitaper_params, brain_areas=''):
    brain_areas = [brain_areas] if isinstance(
    brain_areas, str) else brain_areas
    brain_areas_tetrodes = tetrode_info[tetrode_info.area.isin(brain_areas)]

    lfps = get_LFPs(brain_areas_tetrodes.index, animals)
    start_time = lfps.index.min().total_seconds()

    m = Multitaper(lfps.values, **multitaper_params,
                   start_time=start_time)
    c = Connectivity.from_multitaper(m)
    dimension_names = ['time', 'frequency', 'tetrode']
    data_vars = {
        'power': (dimension_names, c.power())}
    coordinates = {
        'time': _center_time(c.time),
        'frequency': c.frequencies + np.diff(c.frequencies)[0] / 2,
        'tetrode': brain_areas_tetrodes.tetrode_id.values,
        'brain_area': ('tetrode', brain_areas_tetrodes.area.tolist()),
    }
    power = xr.Dataset(data_vars, coords=coordinates).power
    
    return power

multitaper_params = dict(
    sampling_frequency=SAMPLING_FREQUENCY,
    time_window_duration=0.020,
    time_window_step=0.020,
    time_halfbandwidth_product=1,
)

power_CA1 = get_power(tetrode_info, ANIMALS, multitaper_params, brain_areas='ca1')
power_change_CA1 = power_CA1 / power_CA1.mean('time')
ripple_power_CA1 = power_CA1.sel(frequency=slice(150, 250)).mean(['frequency']).interp(dict(time=position_info.index.total_seconds()))
ripple_power_change_CA1 = power_change_CA1.sel(frequency=slice(150, 250)).mean(['frequency', 'tetrode'])

power_MEC = get_power(tetrode_info, ANIMALS, multitaper_params, brain_areas='mec')
power_change_MEC = power_MEC / power_MEC.mean('time')
ripple_power_MEC = power_MEC.sel(frequency=slice(150, 250)).mean(['frequency']).interp(dict(time=position_info.index.total_seconds()))
ripple_power_change_MEC = power_change_MEC.sel(frequency=slice(150, 250)).mean(['frequency', 'tetrode'])
```

```python
from ripple_detection import get_multiunit_population_firing_rate
from src.analysis import get_multiunits, get_multiunit_spikes

multiunit_CA1 = get_multiunits(tetrode_info, brain_areas='ca1')
multiunit_spikes_CA1 = get_multiunit_spikes(multiunit_CA1)
multiunit_MEC = get_multiunits(tetrode_info, brain_areas='mec')
multiunit_spikes_MEC = get_multiunit_spikes(multiunit_MEC)

MUA_CA1 = get_multiunit_population_firing_rate(multiunit_spikes_CA1, SAMPLING_FREQUENCY, 0.015)
MUA_change_CA1 = MUA_CA1 / np.nanmean(MUA_CA1)
MUA_MEC = get_multiunit_population_firing_rate(multiunit_spikes_MEC, SAMPLING_FREQUENCY, 0.015)
MUA_change_MEC = MUA_MEC / np.nanmean(MUA_MEC)
```

```python
from loren_frank_data_processing import reshape_to_segments
from itertools import product

time = position_info.index
times = [hse_times_CA1, ripple_times_CA1, hse_times_MEC, ripple_times_MEC]
times_names = ['High Synchrony Events\nCA1', 'Ripple Events\nCA1', 'High Synchrony Events\nMEC', 'Ripple Events\nMEC']
data = [MUA_change_CA1, ripple_power_change_CA1, MUA_change_MEC, ripple_power_change_MEC]
data_names = ['MUA Change\nCA1', 'Ripple Power Change\nCA1', 'MUA Change\nMEC', 'Ripple Power Change\nMEC']

fig, axes = plt.subplots(4, 4, figsize=(12, 12), sharex=True, sharey=True)

for ax, (datum, times), (datum_name, times_name) in zip(axes.flat, product(data, times), product(data_names, times_names)):
    if isinstance(datum, np.ndarray):
        datum = pd.DataFrame(datum, index=position_info.index.rename('time'))
    elif isinstance(datum, xr.DataArray):
        datum = datum.to_dataframe()
        datum = datum.set_index(pd.TimedeltaIndex(datum.index, unit='s', name='time'))
        time_index = np.digitize(datum.index.total_seconds(),
                                 time.total_seconds())
        time_index[time_index >= len(time)] = len(time) - 1
        datum = (datum.groupby(time[time_index]).mean()
                 .reindex(index=time).interpolate())

    time_locked_datum = reshape_to_segments(
        datum, times, window_offset=(-0.5, 0.5), sampling_frequency=SAMPLING_FREQUENCY)
    t = time_locked_datum.unstack(level=0).index.total_seconds()
    average = np.squeeze(time_locked_datum.mean(level='time').values)
    n_events = time_locked_datum.unstack().shape[0]
    sem = np.squeeze(time_locked_datum.std(level='time').values / np.sqrt(n_events))
    ci_lo, ci_hi = average - sem, average + sem
    ax.plot(t, average, linewidth=2)
    ax.fill_between(t, ci_lo, ci_hi, alpha=0.3)
    ax.axvline(0, color='black', linestyle='--', zorder=-1)
    ax.axhline(np.nanmean(datum), color='black', linestyle='--', zorder=-1)
    ax.set_ylabel(datum_name)
    ax.set_title(times_name)

plt.tight_layout()
```

```python
from src.analysis import get_event_indicator

time = position_info.index
hse_indicator_CA1 = get_event_indicator(time, hse_times_CA1)
hse_indicator_MEC = get_event_indicator(time, hse_times_MEC)
ripple_indicator_CA1 = get_event_indicator(time, ripple_times_CA1)
ripple_indicator_MEC = get_event_indicator(time, ripple_times_MEC)
```

```python
from loren_frank_data_processing import get_all_spike_indicators, make_neuron_dataframe

neuron_info = make_neuron_dataframe(ANIMALS).xs(epoch_key, drop_level=False)
neuron_info = (pd.merge(neuron_info.reset_index('neuron_number'),
                        tetrode_info, left_index=True, right_index=True)
               .set_index('neuron_number', drop=False, append=True))
spikes_CA1 = get_all_spike_indicators(neuron_info[neuron_info.area == 'ca1'].index, ANIMALS)
spikes_MEC = get_all_spike_indicators(neuron_info[neuron_info.area == 'mec'].index, ANIMALS)
```

```python
from replay_identification import ReplayDetector
from sklearn.neighbors import KernelDensity

detector_CA1 = ReplayDetector(
    replay_speed=16,
    spike_model_knot_spacing=10,
    multiunit_density_model=KernelDensity,
    multiunit_model_kwargs=dict(bandwidth=10, leaf_size=100000, rtol=1E-1))
detector_CA1.fit(
    is_replay=ripple_indicator_CA1.values,
    speed=position_info.speed.values,
    position=position_info.linear_distance.values,
    lfp_power=ripple_power_CA1.values,
    spikes=spikes_CA1.values,
    multiunit=multiunit_CA1[:, :4, :],
)
```

```python
detector_CA1.plot_fitted_place_fields(sampling_frequency=SAMPLING_FREQUENCY)
```

```python
results_spikes = detector_CA1.predict(
    speed=position_info.speed.values,
    position=position_info.linear_distance.values,
    spikes=spikes_CA1.values,
    time=position_info.index.total_seconds().values,
    use_likelihoods=['spikes'])

fig, axes = plt.subplots(2, 1, sharex=True, figsize=(15, 6))
axes[0].plot(results_spikes.time, results_spikes.replay_probability)
axes[0].set_title('Replay Probability')
axes[1].plot(results_spikes.time, position_info.linear_distance.values)
axes[1].set_title('Linear Distance')
axes[-1].set_xlabel('Time');
```

```python
import scipy.ndimage
from ipywidgets import interact

time_slices = scipy.ndimage.find_objects(*scipy.ndimage.label(results_spikes.replay_probability > 0.3))


def plot_time_slice(time_slice_ind):
    try:
        plt.figure()
        index = time_slices[time_slice_ind][0]
        g = results_spikes.replay_posterior.isel(time=index).plot(x='time', y='position')
        p = position_info.iloc[index]
        g.axes.plot(p.index.total_seconds(), p.linear_distance.values, linestyle='--', color='white')
    except ValueError:
        pass

interact(plot_time_slice, time_slice_ind=(0, len(time_slices)))
```
